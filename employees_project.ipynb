```python
from faker import Faker
import random
import pandas as pd
from datetime import datetime, timedelta
```

```python
now = datetime.now()
min_date = now - timedelta(days=10)

fake = Faker()
n = 10000

employees = []
for _ in range(n):
    employee = {
        'id': _,
        'name': fake.name(),
        'email': fake.email(),
        'telephone': random.randint(9000000000, 9999999999),
        'address': fake.address(),
        'last_check_in': pd.to_datetime(pd.date_range(min_date, now, periods=1).values[0]),
        'department': fake.random_element(elements=('Sales', 'Marketing', 'IT', 'HR')),
        'salary': random.randint(20000 // 5000, 150000 // 5000) * 5000,
        'age': random.randint(20, 70),
        'managers_id': None,
        'position': None,
        'relatives_id': random.randint(0, n) if fake.boolean(chance_of_getting_true=20) else None,
        'start_date': str(fake.date_between(start_date='-10y', end_date='today')),
        'end_date': str(fake.date_between(start_date='-5y', end_date='today')) if fake.boolean(chance_of_getting_true=5) else None
    }
    employees.append(employee)
```

```python
df = pd.DataFrame.from_dict(employees)
```

```python
for department, dep_data in dict_hierarchy.items():
  for employee in dep_data:
    df.loc[df['id']==employee[1], 'managers_id'] = employee[2]
    df.loc[df['id']==employee[1], 'department'] = department
    df.loc[df['id']==employee[1], 'position'] = employee[0]
```

```python
marketing_managers = set(df[(df['position'] == 'Manager') & (df['department'] == 'Marketing')]['id'].tolist())
sales_managers = set(df[(df['position'] == 'Manager') & (df['department'] == 'Sales')]['id'].tolist())
hr_managers = set(df[(df['position'] == 'Manager') & (df['department'] == 'HR')]['id'].tolist())
it_managers = set(df[(df['position'] == 'Manager') & (df['department'] == 'IT')]['id'].tolist())
```

```python
employees = []
for i in range(len(df)):
  if (df['department'][i] == 'Marketing') and (df['managers_id'][i] == None):
    employees.append(fake.random_element(elements=marketing_managers))
  elif (df['department'][i] == 'Sales') and (pd.isna(df['managers_id'][i])):
    employees.append(fake.random_element(elements=sales_managers))
  elif (df['department'][i] == 'HR') and (pd.isna(df['managers_id'][i])):
    employees.append(fake.random_element(elements=hr_managers))
  elif (df['department'][i] == 'IT') and (pd.isna(df['managers_id'][i])):
    employees.append(fake.random_element(elements=it_managers))
  else:
    employees.append(df['managers_id'][i])
```

```python
df['managers_id'] = employees
```

```python
df.loc[df['position'].isna(), 'position'] = 'Specialist'
```
